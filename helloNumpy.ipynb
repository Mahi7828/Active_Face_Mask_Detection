```python
# Import necessary libraries
from PIL import Image  # Pillow for image manipulation
import numpy as np  # Numpy for array operations

# Load the image 
image = Image.open("the-swing.webp")

# Convert the image to a gray scale and then to a numpy array and print first 5 rows
# Print the shape of the array

# Convert the image to grayscale
gray_image = image.convert("L")

# Convert the grayscale image to a numpy array
gray_image_array = np.array(gray_image)

# Print the first 5 rows of the array
print(gray_image_array[:5])

# Print the shape of the array
print(gray_image_array.shape)

```

### Various Numpy operations

```python
# 1. Reshape and flatten the above array from 2D to 1D and print first 10 values
# Reshape the array to 1D
gray_image_array_1d = gray_image_array.flatten()
# Print first 10 values of the array
print(gray_image_array_1d[:10])   



# Reshape the array back to 2D and print first 5 rows
gray_image_array_2d = gray_image_array_1d.reshape(gray_image_array.shape)
# Print first 5 rows of the array
print(gray_image_array_2d[:5])


#---------------------------------------------------------------------------------------------------------------------------
# 2. Slice the array and get the center of the image of dimensions (height//4, width//4) 
height, width = gray_image_array.shape
# Get the center of the image
center = gray_image_array[height//4:3*height//4, width//4:3*width//4]
# Print the shape of the center
print(center.shape)



# Convert the pixels into new image and show that image
# Convert the center to an image
center_image = Image.fromarray(center)
# Show the center image
center_image.show()


#---------------------------------------------------------------------------------------------------------------------------
# 3. Increase the brightness of the image; and print the pixel values before and after increasing
# Increase the brightness of the image
brighter_image = gray_image_array + 50
# Print the pixel values before and after increasing the brightness
print(gray_image_array[:5])
print(brighter_image[:5])

# Show the new brighter image
brighter_image_image = Image.fromarray(brighter_image)
brighter_image_image.show()





```

Using scipy

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

image = np.array(normalized_image)
# Sobel filter for edge detection
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

sobel_y = np.array([[-1, -2, -1],
                    [ 0,  0,  0],
                    [ 1,  2,  1]])

# Apply the Sobel filter to the image using convolve2d
edges_x = convolve2d(image, sobel_x, mode='same', boundary='fill', fillvalue=0)
edges_y = convolve2d(image, sobel_y, mode='same', boundary='fill', fillvalue=0)

# Combine the edges
edges = np.hypot(edges_x, edges_y)

# Plot the original image and the edge-detected image
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Edge Detected Image')
plt.imshow(edges, cmap='gray')

plt.show()
```

Using numpy

```python
import numpy as np
import matplotlib.pyplot as plt

# Example image (you can replace this with your own image)
image = np.array(normalized_image)

# Sobel filter for edge detection
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

sobel_y = np.array([[-1, -2, -1],
                    [ 0,  0,  0],
                    [ 1,  2,  1]])

def convolve2d(image, kernel):
    kernel_height, kernel_width = kernel.shape
    image_height, image_width = image.shape
    
    # Calculate the padding width and height
    pad_height = kernel_height // 2
    pad_width = kernel_width // 2
    
    # Pad the image with zeros
    padded_image = np.pad(image, ((pad_height, pad_height), (pad_width, pad_width)), mode='constant')
    
    # Initialize the output image
    output = np.zeros_like(image)
    
    # Perform the convolution
    for i in range(image_height):
        for j in range(image_width):
            region = padded_image[i:i + kernel_height, j:j + kernel_width]
            output[i, j] = np.sum(region * kernel)
    
    return output

# Apply the Sobel filter to the image
edges_x = convolve2d(image, sobel_x)
edges_y = convolve2d(image, sobel_y)

# Combine the edges
edges = np.hypot(edges_x, edges_y)



# Plot the original image and the edge-detected image
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Edge Detected Image')
plt.imshow(edges, cmap='gray')

plt.show()
```
